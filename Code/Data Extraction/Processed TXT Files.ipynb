```python
path = '/content/drive/Shareddrives/CrimeTrack: Tracking Crime on the Web/Data/Tweets Guardia Civil.txt'
```

```python
tweets_policia = []
with open(path) as f:
    for line in f:
        for tweet in jsonpickle.decode(line).values():
            tweets_policia.append(json.loads(tweet))
```

```python
df = pd.DataFrame(index=range(64132),columns = ['username', 'creation_date', 'tweet', 'likes', 'retweets'])
for i in range(len(tweets_policia)): 
    if 'referenced_tweets' in tweets_policia[i].keys() and 'text' in tweets_policia[i]['referenced_tweets'][0].keys():  
        df['username'][i] = tweets_policia[i]['author']['username'] 
        df['creation_date'][i] = tweets_policia[i]['created_at'] 
        df['tweet'][i] = tweets_policia[i]['referenced_tweets'][0]['text']
        df['likes'][i] = tweets_policia[i]['public_metrics']['like_count']
        df['retweets'][i] = tweets_policia[i]['public_metrics']['retweet_count']

    else: 
        df['username'][i] = tweets_policia[i]['author']['username'] 
        df['creation_date'][i] = tweets_policia[i]['created_at'] 
        df['tweet'][i] = tweets_policia[i]['text']
        df['likes'][i] = tweets_policia[i]['public_metrics']['like_count']
        df['retweets'][i] = tweets_policia[i]['public_metrics']['retweet_count']
```

```python
df['creation_date'] = pd.to_datetime(df['creation_date'])
df['creation_date'] = df['creation_date'].dt.tz_localize(None)
```
