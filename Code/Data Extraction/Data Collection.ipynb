```python
from twarc import Twarc2, expansions
import json
```

```python
tweets_guardia = {}
count = 0

client = Twarc2(bearer_token="AAA...........wM1")
for page in client.search_all(query='from:guardiacivil'):
    result = expansions.flatten(page)
    for tweet in result:
        tweets_guardia[count] = json.dumps(tweet)
        count += 1
```

```python
tweets_mossos = {}
count = 0

client = Twarc2(bearer_token="AAA...........wM1")
for page in client.search_all(query='from:mossos lang:ca'):
    result = expansions.flatten(page)
    for tweet in result:
        tweets_mossos[count] = json.dumps(tweet)
        count += 1
```

```python
tweets_policia = {}
count = 0

client = Twarc2(bearer_token="AAA...........wM1")
for page in client.search_all(query='from:policia lang:es'):
    result = expansions.flatten(page)
    for tweet in result:
        tweets_policia[count] = json.dumps(tweet)
        count += 1
```

```python
with open('Tweets Mossos.txt', 'w') as file:
     file.write(json.dumps(tweets_mossos))
```
