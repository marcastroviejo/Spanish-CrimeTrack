```python
import googletrans
from googletrans import Translator
import pandas as pd

translator = Translator()
```

```python
#Mount Google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
import emoji
def give_emoji_free_text(text):
    return emoji.get_emoji_regexp().sub("", text)
```

```python
path = '/content/drive/Shareddrives/CrimeTrack: Tracking Crime on the Web/Data/Train Dataset/train_dataset_catalan_1.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
translator.raise_Exception = True
```

```python
for i in range(len(df)): 
   df['tweet_ca'][i] = translator.translate(df['tweet_processed_es'][i], dest='ca').text
   if (i) % 100 == 0:
        print(i)
```

```python
df['creation_date'] = pd.to_datetime(df['creation_date'])
df['creation_date'] = df['creation_date'].dt.tz_localize(None)
```

```python
df.to_excel('/content/drive/Shareddrives/CrimeTrack: Tracking Crime on the Web/Data/Train Dataset/train_dataset_catalan_1.xlsx', index=False)
```
